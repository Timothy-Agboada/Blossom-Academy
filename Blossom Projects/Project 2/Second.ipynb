```python
import pandas as pd
```

```python
import boto3
```

```python
region = "eu-west-1"
s3 = boto3.client('s3',region_name=region)
#location = {'LocationConstraint': region}
bucket = "blossom-data-engs"

s3.download_file(bucket,"free-7-million-company-dataset.zip",'data.zip')
```

```python
data = pd.read_csv('data.zip',compression = 'zip')
```

```python
data = data.dropna()
```

```python
import pandas as pd
import pyarrow as pa
t = pa.Table.from_pandas(data)
data.to_parquet('free-data-7-million-dataset.parquet')
data.to_json('free-data-7-million-dataset.json.gzip', compression='gzip')
```
